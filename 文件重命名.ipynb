```python
c = Counter(new_name1)  #统计文件名中出现多少次，以此找出同音重复出现的名字
c_plus=list(({k:v for k, v in c.items() if v>=2}).keys())  #选出出现两次以上的名字
c_plus
```

```python
# 确定出现两次以上的名字，他们在列表中的索引位置
b=[]
for n in range(len(c_plus)):
    for index, name in enumerate(new_name1):
        if name == c_plus[n]:
            b.append(index)
print(b)
```
